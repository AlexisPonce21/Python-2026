### **2.- Programa en Python que lea 2 números, realizar la suma y desplegar la suma de los 2 números**

```python
numero1 = int(input("Escriba un numero "))
numero2 = int(input("Escriba otro numero "))
print (f"El resultado de la suma es {numero1 + numero2}")
```

### **3.- Programa en Python que lea 2 números y realice las 4 operaciones básicas**

```python
numero1 = int(input("Escriba un numero "))
numero2 = int(input("Escriba otro numero "))
print ("Los resultados de las operaciones son:")
print (f"Suma = {numero1 + numero2}")
print (f"Resta = {numero1 - numero2}")
print (f"Multiplicacion = {numero1 * numero2}")
print (f"Division = {numero1 / numero2}")
```

### **4.- Programa en Python Que sirva para calcular el área de un triángulo**

```python
base = int(input("Escriba la base del triangulo "))
altura = int(input("Escriba la altura del triangulo "))
print (f"El area del triangulo es {(base * altura)/2}")
```

### **5.- Programa en Python que lea una medida en centímetros y desplegar la misma medida pero convertida en pies y pulgadas**

```python
centimetros = int(input("Escriba la medida en centimetros "))
print ("Las conversiones serian las siguientes:")
print (f"Pies = {centimetros / 30.48}")
print (f"Pulgadas = {centimetros / 2.54}")
```

### **6.- Programa en Python que lea una temperatura en grados centígrados y desplegarla en Grados Fahrenheit y grados kelvin**

```python
centigrados = int(input("Escriba los grados centigrados "))
print ("Las conversiones son las siguientes:")
print (f"Fahrenheit = {(centigrados * 9 / 5) + 32}")
print (f"Kelvin = {centigrados + 273.15}")
```

### **7.- Programa en Python que lea 4 calificaciones de un alumno, desplegar el promedio del alumno**

```python
calif1 = int(input("Escriba la primera calificacion "))
calif2 = int(input("Escriba la segunda calificacion "))
calif3 = int(input("Escriba la tercera calificacion "))
calif4 = int(input("Escriba la cuarta calificacion "))
print (f"El promedio final es de {(calif1 + calif2 + calif3 + calif4)/4}")
```
